```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px

y = ['temp']

df = pd.read_csv("temp_ext_COOP.csv")
fig = px.line(df, x='dt', y=y)

#fig.add_hline(y=28, line_width=1, line_dash="dash", line_color="blue")
#fig.add_hline(y=30, line_width=1, line_dash="dash", line_color="red", annotation_text="Seuil inconfort 28°", 
            #  annotation_position="bottom right",
            #  annotation_font_size=15,
            #  annotation_font_color="blue")

              
fig.update_layout(xaxis_range=['2021-04-27','2021-09-03'],
                  title_text="Température Exterieur Grenoble 2021")
#fig.update_xaxes(rangeslider_visible=True)
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=1, label="1d", step="day", stepmode="backward"),
            dict(count=1, label="1h", step="hour", stepmode="todate"),
            dict(step="all")
        ])
    )
)
fig.show()
```
